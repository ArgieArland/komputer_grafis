```python
import matplotlib.pyplot as plt
import numpy as np

plt.axis([0, 140, 0, 100])
plt.grid(True)

t = [20, 40, 60, 80, 100, 120]     # time
T = [30, 35, 43, 55, 70, 85]       # temperature
p = [2, 3, 4.5, 7.3, 9.6, 11]      # pressure (tambah 1 data biar sama panjang)
v = [0.6, 0.58, 0.54, 0.46, 0.35, 0.2]  # volume

pp = [val * 10 for val in p]   # scale pressure
vv = [val * 100 for val in v]  # scale volume

plt.plot(t, T, color='r', label='Temperature', marker='o')
plt.plot(t, pp, color='b', label='Pressure', marker='s')
plt.plot(t, vv, color='g', label='Volume', marker='d')

plt.legend(loc='upper left')

# Plot pressure scale values
for y in np.arange(0, 101, 20):
    plt.text(130, y, f"{y/10:.1f}", color='b')  # tampilkan pressure (psi)

# Plot volume scale values
for y in np.arange(0, 101, 20):
    plt.text(135, y, f"{y/100:.2f}", color='g')  # tampilkan volume (cm3)

plt.xlabel('Time (hrs)')
plt.ylabel('Temperature (K)', color='r')
plt.text(132, 80, 'Pressure (psi)', rotation=90, color='b')
plt.text(137, 80, 'Volume (cm3)', rotation=90, color='g')

plt.title('Compression Test Results')
plt.show()
```

```python
"""
REGRESSION1
"""
#--------same as DATAPLOT5
import matplotlib.pyplot as plt
import numpy as np

plt.axis([0, 140, 0, 100])
plt.grid(True)

t = [20, 40, 60, 80, 100, 120]
T = [30, 35, 43, 55, 70, 85]
p = [2, 3, 4.5, 7.3, 9.6, 11]
v = [0.6, 0.58, 0.54, 0.46, 0.35, 0.2]

pp = [val * 10 for val in p]
vv = [val * 100 for val in v]

plt.plot(t, T, color='r', label='Temperature', marker='o')
plt.plot(t, pp, color='b', label='Pressure', marker='s')
plt.plot(t, vv, color='g', label='Volume', marker='d')
plt.legend(loc='upper left')

# Plot pressure scale values
for y in np.arange(0, 101, 20):
    plt.text(130, y, f"{y/10:.1f}", color='b')

# Plot volume scale values
for y in np.arange(0, 101, 20):
    plt.text(135, y, f"{y/100:.2f}", color='g')

plt.xlabel('Time (hrs)')
plt.ylabel('Temperature (K)', color='r')
plt.text(132, 80, 'Pressure (psi)', rotation=90, color='b')
plt.text(137, 80, 'Volume (cm3)', rotation=90, color='g')

plt.title('Compression Test Results')

# REGRESSION LINE for v vs t
n = len(v)
c1 = np.sum(t) / n
c2 = np.sum(v) / n
vt = np.multiply(v, t)
tt = np.multiply(t, t)
c3 = np.sum(vt)
c4 = np.sum(tt)

A = (c3 - n * c1 * c2) / (c4 - n * c1 * c1)
B = c2 - A * c1

# Scatter plot of regression line points
for tp in np.arange(t[0], t[-1]+1, 2):
    vp = A * tp + B
    vp_scaled = vp * 100
    plt.scatter(tp, vp_scaled, color='g', s=10)

# RMS error
sumee = 0
for i in range(n):
    ee = v[i] - (A * t[i] + B)
    sumee += ee * ee
rms = np.sqrt(sumee / n)

# Labels
plt.text(60, 28, 'v=At+B', color='g')
plt.arrow(78, 30, 6, 6, head_length=3, head_width=1.5, color='g', linewidth=.5)

vp1 = A * t[0] + B
plt.text(2, 64, f"{vp1:.4f}", color='g')

vp2 = A * t[-1] + B
plt.text(122, 25, f"{vp2:.4f}", color='g')

plt.text(65, 18, 'A=', color='g')
plt.text(72, 18, f"{A:.5f}", color='g')

plt.text(65, 12, 'B=', color='g')
plt.text(73, 12, f"{B:.5f}", color='g')

plt.text(95, 3, 'RMS error=', color='g')
plt.text(123, 3, f"{rms:.3f}", color='g')

plt.show()
```
