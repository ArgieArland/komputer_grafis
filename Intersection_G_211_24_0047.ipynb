```python
"""
LRP
"""

import numpy as np
import matplotlib.pyplot as plt
from math import sin, cos, radians, sqrt

# --------- koordinat awal
xg = []
yg = []
zg = []

xc = 80
yc = 40
zc = 40

x = [-40, -40, 40, 40, -40, 50]
y = [0, 0, 0, 0, 0, 3]
z = [-10, 10, 10, -10, 15, -10]

for i in range(len(x)):
    xg.append(x[i] + xc)
    yg.append(y[i] + yc)
    zg.append(z[i] + zc)

# --------- fungsi rotasi
def rotx(xc, yc, zc, xp, yp, zp, Rx):
    y_shift = yp - yc
    z_shift = zp - zc
    y_new = y_shift * cos(Rx) - z_shift * sin(Rx) + yc
    z_new = y_shift * sin(Rx) + z_shift * cos(Rx) + zc
    return xp, y_new, z_new

def roty(xc, yc, zc, xp, yp, zp, Ry):
    x_shift = xp - xc
    z_shift = zp - zc
    x_new = x_shift * cos(Ry) + z_shift * sin(Ry) + xc
    z_new = -x_shift * sin(Ry) + z_shift * cos(Ry) + zc
    return x_new, yp, z_new

def rotz(xc, yc, zc, xp, yp, zp, Rz):
    x_shift = xp - xc
    y_shift = yp - yc
    x_new = x_shift * cos(Rz) - y_shift * sin(Rz) + xc
    y_new = x_shift * sin(Rz) + y_shift * cos(Rz) + yc
    return x_new, y_new, zp

# --------- fungsi hit point
def hitpoint(x, y, z):
    a = x[5] - x[4]
    b = y[5] - y[4]
    c = z[5] - z[4]
    Q45 = sqrt(a * a + b * b + c * c)

    lx = a / Q45
    ly = b / Q45
    lz = c / Q45

    a = x[3] - x[0]
    b = y[3] - y[0]
    c = z[3] - z[0]
    Q03 = sqrt(a * a + b * b + c * c)

    ux = a / Q03
    uy = b / Q03
    uz = c / Q03

    a = x[1] - x[0]
    b = y[1] - y[0]
    c = z[1] - z[0]
    Q01 = sqrt(a * a + b * b + c * c)

    vx = a / Q01
    vy = b / Q01
    vz = c / Q01

    nx = uy * vz - uz * vy
    ny = uz * vx - ux * vz
    nz = ux * vy - uy * vx

    vxl = x[4] - x[0]
    vyl = y[4] - y[0]
    vzl = z[4] - z[0]

    Qm = (vxl * nx + vyl * ny + vzl * nz)

    cosp = lx * nx + ly * ny + lz * nz
    Qh = abs(Qm / cosp)

    xh = x[4] + Qh * lx
    yh = y[4] + Qh * ly
    zh = z[4] + Qh * lz

    xhg = xh + xc
    yhg = yh + yc
    zhg = zh + zc

    # --- check bounds
    a = xh - x[0]
    b = yh - y[0]
    c = zh - z[0]

    up = a * ux + b * uy + c * uz
    vp = a * vx + b * vy + c * vz

    hitcolor = 'r'
    if up < 0 or up > Q03 or vp < 0 or vp > Q01:
        hitcolor = 'b'

    if Q45 < Qh:
        hitcolor = 'g'

    return xh, yh, zhg, xhg, yhg, hitcolor

# --------- plot sistem
def plotsystem(xg, yg, zg, xh, yh, zhg, hitcolor):
    plt.plot([xg[0], xg[1], xg[2], xg[3], xg[0]],
             [yg[0], yg[1], yg[2], yg[3], yg[0]], color='k')
    plt.plot([xg[4], xg[5]], [yg[4], yg[5]], color='b')

    if hitcolor == 'g':
        plt.scatter(xg[5], yg[5], s=20, color=hitcolor)
    else:
        plt.scatter(xh, yh, s=20, color=hitcolor)

    plt.axis([0, 150, 0, 100])
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

# --------- rotasi + plot
def plotx(xc, yc, zc, Rx):
    for i in range(len(x)):
        xg[i], yg[i], zg[i] = rotx(xc, yc, zc, xg[i], yg[i], zg[i], Rx)
        x[i] = xg[i] - xc
        y[i] = yg[i] - yc
        z[i] = zg[i] - zc

    xh, yh, zhg, xhg, yhg, hitcolor = hitpoint(x, y, z)
    plotsystem(xg, yg, zg, xhg, yhg, zhg, hitcolor)

def ploty(xc, yc, zc, Ry):
    for i in range(len(x)):
        xg[i], yg[i], zg[i] = roty(xc, yc, zc, xg[i], yg[i], zg[i], Ry)
        x[i] = xg[i] - xc
        y[i] = yg[i] - yc
        z[i] = zg[i] - zc

    xh, yh, zhg, xhg, yhg, hitcolor = hitpoint(x, y, z)
    plotsystem(xg, yg, zg, xhg, yhg, zhg, hitcolor)

def plotz(xc, yc, zc, Rz):
    for i in range(len(x)):
        xg[i], yg[i], zg[i] = rotz(xc, yc, zc, xg[i], yg[i], zg[i], Rz)
        x[i] = xg[i] - xc
        y[i] = yg[i] - yc
        z[i] = zg[i] - zc

    xh, yh, zhg, xhg, yhg, hitcolor = hitpoint(x, y, z)
    plotsystem(xg, yg, zg, xhg, yhg, zhg, hitcolor)

# --------- loop input
while True:
    axis = input('x, y or z? (enter to quit): ')
    if axis == 'x':
        Rx = radians(float(input('Rx Degrees?: ')))
        plotx(xc, yc, zc, Rx)
    elif axis == 'y':
        Ry = radians(float(input('Ry Degrees?: ')))
        ploty(xc, yc, zc, Ry)
    elif axis == 'z':
        Rz = radians(float(input('Rz Degrees?: ')))
        plotz(xc, yc, zc, Rz)
    elif axis == '':
        break
```
